```python
Customers = pd.DataFrame(data, columns = ['Name', 'Address', 'Turnover', 'Assets', 'Withdrawal', 
                                          'Deposit', 'Student', 'Retired', 'Employed', 'Resident'])
```

```python
#Adding '+' instead of space as google combine all string together
Customers['Address']=Customers['Address'].str.replace(' ', '+')
```

```python
#As attributes are objects, requires to change it into strings
Customers['Address'] = Customers['Address'].astype(str)
```

```python
#Adding new columns for URLs
Customers['Url'] = ['https://www.google.com/maps/search/' + i for i in Customers['Address'] ]
```

```python
CustomersUrl_With_Coordinates = []

option = webdriver.ChromeOptions()
prefs = {'profile.default_content_setting_values': {'images':2, 'javascript':2}}
option.add_experimental_option('prefs', prefs)

driver = webdriver.Chrome("/Users/amihailovs/Desktop/MasterThesis/chromedriver", options=option)

for url in tqdmn(Customers['Url'], leave=False):
    driver.get(url)
    CustomersUrl_With_Coordinates.append(driver.find_element_by_css_selector('meta[itemprop=image]').get_attribute('content'))
    
driver.close()
```

```python
Customers['Url_With_Coordinates'] = CustomersUrl_With_Coordinates
```
